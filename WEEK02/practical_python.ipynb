```python
# ==========================================================
# FIX ME CHALLENGE: PROFIL PERMULAAN ROBOT
# ==========================================================

robot_name = "KV-TRON Safety Patrol"          # Contoh: "KV-TRON Safety Patrol"
robot_id = 295                # Contoh: 205
battery_level = 95           # Contoh: 95
operation_mode = "Pemeriksaan Bengkel"      # Contoh: "Pemeriksaan Bengkel"
robot_location = "Bengkel TVET A"      # Contoh: "Bengkel TVET A"

# FIX ME: Tukar kepada input("Masukkan nama jurutera: ")
engineer_name = input("Masukkan nama jurutera: ")

print("\n====================================")
print("       PROFIL PERMULAAN ROBOT")
print("====================================")
print("Nama Robot    :", robot_name)
print("ID Robot      :", robot_id)
print("Bateri        :", battery_level, "%")
print("Mod Operasi   :", operation_mode)
print("Lokasi        :", robot_location)
print("Jurutera      :", engineer_name)

lengkap = (
    robot_name != "robot_name"
    and robot_id is not None
    and battery_level is not None
    and operation_mode != "operation_mode"
    and robot_location != "robot_location"
    and engineer_name != "engineer_name"
)

print("------------------------------------")
print("[ROBOT ONLINE] Identiti lengkap." if lengkap
      else "[BELUM LENGKAP] Baiki semua bahagian FIX ME.")
```

#Latihan Praktikal: Sistem Pendaftaran Ahli Kelab Robotik
Situasi

Anda merupakan seorang pembantu makmal AI di Kolej Vokasional. Hari ini anda diminta membina satu program ringkas untuk mendaftarkan ahli baharu Kelab Robotik.

Program anda perlu mendapatkan maklumat daripada pengguna dan memaparkannya semula dengan kemas.

Tugasan

Bina satu program yang boleh:

Meminta pengguna memasukkan:
Nama

*   Umur
*   Program Pengajian

*   Semester
*   Nama Robot Pilihan

Paparkan semua maklumat tersebut dengan format yang kemas.
Paparkan jenis data (data type) bagi setiap maklumat yang dimasukkan.

```python
print("========================================")
print("     PENDAFTARAN AHLI KELAB ROBOTIK     ")
print("========================================")
print()

nama = input("Nama          : ")
umur = input("Umur          : ")
program = input("Program       : ")
semester = input("Semester      : ")
robot = input("Robot Pilihan : ")

tahun_pengajian = input("Tahun Pengajian: ")
minat_ai = input("Minat dalam AI : ")

print()
print("========================================")
print("Jenis Data")
print("========================================")

print("Nama          :", type(nama))
print("Umur          :", type(umur))
print("Program       :", type(program))
print("Semester      :", type(semester))
print("Robot Pilihan :", type(robot))

print("Tahun Pengajian:", type(tahun_pengajian))
print("Minat dalam AI :", type(minat_ai))
```
